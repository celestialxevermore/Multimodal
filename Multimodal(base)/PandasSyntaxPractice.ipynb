### DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

#### 1) 리스트 ,ndarray에서 DataFrame으로 변환

```python
import numpy as np

col_name1=['col1']
list1 = [1,2,3]

array1 = np.array(list1)
#리스트 객체를 ndarray로 변환한 결과

print('array1 shape:',array1.shape)

df_list1 = pd.DataFrame(list1, columns = col_name1)
#데이터프레임으로의 변환을 list로 하였음

print('1차원 리스트로 만든 DataFrame: \n', df_list1)

df_array1 = pd.DataFrame(array1,columns = col_name1)
print('1차원 ndarray로 만든 DataFrame: \n',df_array1)


```

정리 : ndarray, list 객체 모두 data = ndarray, data = list를 통해 DataFrame으로 만들 수 있다.

```python
#3개의 경우 어떻게 될까?

col_name2 = ['col1','col2','col3']

#2행 3열 형태의 리스트와 ndarray를 생성한 뒤, 이를 DataFrame으로 변환하겠습니다.
list2=[]
for i in range(2):
    tmp=[]
    if i == 0:
        for j in range(1,4):
            tmp.append(j)
        
    elif i==1:
        for j in range(1,4):
            tmp.append(i*10+j)
        
    list2.append(tmp)
        
list2

array2 = np.array(list2)
array2
print('array2 shape:',array2.shape)

df_list2 = pd.DataFrame(data = list2, columns = col_name2)
print('2차원 리스트로 만든 데이터 프레임:',df_list2)

df_array2 = pd.DataFrame(data = array2,columns=col_name2)
print('2차원 ndarray로 만든 데이터 프레임:',df_array2)
```

#### reset_index() : 인덱스를 연속 숫자형으로 할당하고, 쉽게 말해 index와 똑같은 것을 하나 추가한다.

```python
print('### before reset index ###')
value_counts = titanic_df['Pclass'].value_counts()

#Series 객체로 만들어야 value_counts로 만들 수 있다.

print('value_counts 객체 변수 타입 :',type(value_counts))

new_value_counts = value_counts.reset_index(inplace=False)

print('### After reset_index ###')

print(new_value_counts)

print('new_value_counts 객체 변수 타입 : ',type(new_value_counts))
```

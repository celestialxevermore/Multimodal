#### 텍스트 데이터를 Python string methods를 사용하여 단어 단위로 파싱

```python
#import modules

import numpy as np
import os
```

```python
base_dir = 'C:/Users/user/Jupyter_project/Multimodal_project'
file_name='python_wikipedia.txt'
path = os.path.join(base_dir,file_name)

# Open file and print it as an example

file_opened = open(path)

for line in file_opened.readlines():
    print(line)
```

#### 단어를 정제하기

- 대문자를 소문자로 전부 바꾸기
- stop words를 제거하기.
- 기호 제거하기.
- 숫자 제거하기.

```python
def word_preprocess(word):
    #lower_case
    
    word = word.lower()
    
    #remove stop-words
    #list up the stop words
    stop_words_list = ['a','an','in','with','to','for','from','of','at','until','by','and'
                       'but','is','are','was','were','it','my','his','her','our','as','not']
    
    
    for stopword in stop_words_list:
        if word!=stopword:
            pass
        else:
            word=''
    
    
    # remove symbols such as comma, period, etc.
    symbols = [',','.',':','-','+','/','*','&','%','[', ']','(',')']
    
    for symbol in symbols:
        word = word.replace(symbol,'')
        
    
    # remove numbers
    if word.isnumeric():
        word=''
    
    return word
```

#### 다음 작업.

- 파일을 열어서 한 줄 단위로 읽고(readline)
- 좌우 공백을 제거한 다음, 단어 단위로 분할(split)
- 위에서 정의한 word_preprocessing 함수를 적용하여 싹 전처리를 한 뒤
- token_idx 사전에 단어를 key, Index를 Value로 저장합니다. 

#### Python strip() method

- whitespace : 띄어쓰기 / 엔터 줄바꿈, 탭 을 전부 말합니다.
- strip 메소드는 이러한 whitespace를 없애는 기능을 합니다.
- 어떤 한 라인에 대해서, 맨 앞과 맨 뒤에 존재하는 whitespace를 없앱니다.
- 하지만, 그 라인 안에 있는 whitespace는 없앨 수 없습니다.

```python
file_opened = open(path)

#비어 있는 딕셔너리 객체를 생성합니다.
Frequency_idx = {}


#한 줄 씩 읽어나갑니다.

for line in file_opened.readlines():
    #strip leading and trailing edge spaces
    line = line.strip()
    
    #split the line into word with a space delimiter
    for word in line.split():
        word = word_preprocess(word)
        
        #put the preprocessed word into Frequency_idx
        if word not in Frequency_idx:
            if word !='':
                Frequency_idx[word] = len(Frequency_idx)+1
```

#### 텍스트를 단어 단위로 One-hot encoding 하기

- 하나의 텍스트 라인에서 고려할 최대 단어 빈도를 40으로 정합니다.
- 한 문장에서 41번째부터 나오는 단어는 무시
- 그리고 One-hot encoding한 결과를 저장할 빈 one_hot_encoded 다차원 배열을
- np.zeros()로 만들어두었습니다.

```python
max_len = 40

file_opened = open(path)

one_hot_encoded = np.zeros(shape=(len(file_opened.readlines()),
                                  max_len,
                                  max(Frequency_idx.values())+1))
```

```python
one_hot_encoded.shape
```

- one_hot_encoded는 (5,40,124)의 다차원 배열입니다. 
- 5개의 텍스트 문장으로 되어 있고, 40개의 최대 단어 길이만을 고려합니다.
- 그리고 총 128개의 token index에 대해서 해당 단어가 있다면, 1 / 없으면 0으로 one_hot_encoding을 하게 됩니다.

#### 다음 작업

- 파일을 읽습니다. 텍스트를 줄 단위로 읽어 들인 후에
- for loop를 돌면서 각 줄에서 단어를 분할하고, 전처리까지하고,
- Frequency_idx.get(word) 메소드를 통해 해당 단어의 index를 가져온 다음
- 해당 텍스트(line_num), 단어(word), Frequency_index(idx)에 '1'을 입력하여
- one_hot_encoded 다차원 배열을 Update합니다.

```python
#file open
file_opened = open(path)

for line_num,line in enumerate(file_opened.readlines()):
    
    # strip leading and trailing edge spaces
    # delete whitespace in a line
    line = line.strip()
    
    
    for wordnum,word in list(enumerate(line.split()))[:max_len]:
        
        #preprocess the word
        word = word_preprocess(word)
        
        
        # put word into token_index
        
        if word!='':
            idx = Frequency_idx.get(word)
            one_hot_encoded[line_num,wordnum,idx]=1.
```

#### 3-2) 상황 설명
- 전체 json을 DF로 바꾼 뒤 결측치 제거
- DF['data']가 복잡하니까 이것을 frame / frame 데이터로 나눈것임
- get_frame_info는 그 데이터의 frame 정보만을 담은 리스트임 str형이 들어가 있음
- get_frame_data는 그 데이터의 frame 데이터만을 담은 리스트임 dict형이 들어가 있음
- 모두 array가 아닌 list를 반환함
- 요구에 따라 array로 바꿔서 사용할 수 있음

```python
def get_frame_info(json):
    df2 = pd.DataFrame(json)
    df2 = df2[['data']]

    #filteredDF2
    #print('결측치 제거 전 전체 크기 : ',len(filteredDF2))
    df2 = df2.dropna(axis=0)
    #filteredDF2
    #print('결측치 제거 후 전체 크기 : ',len(filteredDF2))
    df2dict = df2.to_dict()
    df2dict_dict = list(df2dict.values())
    
    frameinfolist = list(df2dict_dict[0].keys())
    return frameinfolist
    
```

```python
def get_frame_data(json):
    df2 = pd.DataFrame(json)
    df2 = df2[['data']]

    #filteredDF2
    #print('결측치 제거 전 전체 크기 : ',len(filteredDF2))
    df2 = df2.dropna(axis=0)
    #filteredDF2
    #print('결측치 제거 후 전체 크기 : ',len(filteredDF2))
    df2dict = df2.to_dict()
    df2dict_dict = list(df2dict.values())
    
    
    framedatalist = list(df2dict_dict[0].values())
    return framedatalist
```

#### 4-1) get_data column_list : 중요!! 원본 데이터프레임에서 깊이가 깊은 data 컬럼의 수많은 네스티드 구조를 전부 컬럼으로 만들고, 하나의 데이터프레임을 반환하는 함수

- 그러니까, 처음부터 이야기 하자면, 전체 json 파일에서 긁어낸 데이터프레임 중, data 컬럼만 띡 뽑아내고
- 모든 frame(1143,241,1167...)별로 싹 전부 메타 정보를 한눈에 볼 수 있게 정리해주는 함수입니다.

```python
def get_data_column_list(json):
    framedatalist = get_frame_data(json)
    ## 맨 처음 0번째 1143 frame의 데이터프레임
    framedataDF0 = pd.DataFrame(framedatalist[0])
    framedataDF0= framedataDF0.transpose()
    
    Total_Count_Of_Frame = len(framedatalist)
    
    for i in range(1,Total_Count_Of_Frame):
        framedatanext = pd.DataFrame(framedatalist[i])
        framedatanext = framedatanext.transpose()
        framedataDF0 = framedataDF0.append(framedatanext)
    
    return framedataDF0
```
